## Node Responsibilities

- `planner.py`: choose strategy and thresholds
- `collector.py`: collect GitHub and RSS items
- `analyzer.py`: analyze one item at a time with the LLM client
- `reviewer.py`: score analyses across five weighted dimensions
- `reviser.py`: rewrite analyses from reviewer feedback
- `organizer.py`: filter, deduplicate, and persist articles
- `human_flag.py`: preserve failed review batches for manual follow-up
